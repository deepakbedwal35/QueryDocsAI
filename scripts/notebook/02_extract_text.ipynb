```python
os.makedirs(text_folder_path, exist_ok=True)

for file_name in os.listdir(pdf_folder_path):
    if file_name.endswith(".pdf"):
        pdf_path = os.path.join(pdf_folder_path, file_name)
        try:
            doc = fitz.open(pdf_path)
            full_paper_text = []
            for page in doc:
                full_paper_text.append(page.get_text())
            doc.close()
            
            raw_text = '\n'.join(full_paper_text)
            text_file_name = file_name.replace(".pdf",".txt")
            text_file_path = os.path.join(text_folder_path, text_file_name)
            with open(text_file_path, "w", encoding="utf-8") as txt_file:
                txt_file.write(raw_text)
                
            print(f"Successfully processed and saved: {text_file_name}")
        
        except Exception as e:
            print(f"Severe error extracting text from {file_name}: {e}")
```

```python

os.makedirs(jsonl_folder_path, exist_ok=True)
splitter = RecursiveCharacterTextSplitter(chunk_size=1200, chunk_overlap=250)

for file_name in os.listdir(text_folder_path):
    if file_name.endswith(".txt"):
        text_file_path = os.path.join(text_folder_path, file_name)
        try:
            with open(text_file_path, "r", encoding="utf-8") as f:
                text_content = f.read()

            chunks = splitter.split_text(text_content)
            jsonl_file_name = file_name.replace(".txt", ".jsonl")
            jsonl_file_path = os.path.join(jsonl_folder_path, jsonl_file_name)

            with open(jsonl_file_path, "a", encoding="utf-8") as jsonl_file:
                for chunk_index, chunk in enumerate(chunks):
                    chunk_payload = {
                        "chunk_id": f"{jsonl_file_name.replace('.jsonl', '')}_c{chunk_index}",
                        "text": chunk,
                    }
                    json_string = json.dumps(chunk_payload, ensure_ascii=False)
                    jsonl_file.write(json_string + "\n")

            print(f"Successfully processed and saved: {jsonl_file_name}")

        except Exception as e:
            print(
                f"Severe error during converting text file in chunks from {file_name}: {e}"
            )
```
